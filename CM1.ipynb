### Normalization

In this part, we are going to see the result of minmax and zscore.

#### Minmax normalization

First, we have defined minmax as follow:

```python
def minmax_scaling(a):
    xx=a
    a_std = (xx- xx.min(axis = 0)) / (xx.max(axis = 0) - xx.min(axis = 0))
    return a_std
```

```python
mf_iris = df_iris_org.copy()
mf_iris['sepal_length']=minmax_scaling(mf_iris['sepal_length'])
mf_iris['sepal_width']=minmax_scaling(mf_iris['sepal_width'])
mf_iris['petal_length']=minmax_scaling(mf_iris['petal_length'])
mf_iris['petal_width']=minmax_scaling(mf_iris['petal_width'])
print('Table5 describing mean, standard deviation, min, max and so on in Iris data base after minmax normalization.')
mf_iris.describe()
```

Based on table 5, the standard deviation decreased significantly. It means that the distribution of data base is between zero and one and we have suppressed big numbers in ths database. it can be considered as a noise suppression method. It is clear that minimum is fixed to zero, and maximum is fixed to 1. 

```python
mf=df_org.copy()
mf['age']=minmax_scaling(mf['age'])
mf['sex']=minmax_scaling(mf['sex'])
mf['trestbps']=minmax_scaling(mf['trestbps'])
mf['chol']=minmax_scaling(mf['chol'])
mf['fbs']=minmax_scaling(mf['fbs'])
mf['thalach']=minmax_scaling(mf['thalach'])
mf['exang']=minmax_scaling(mf['exang'])
mf['oldpeak']=minmax_scaling(mf['oldpeak'])
mf['ca']=minmax_scaling(mf['ca'])
mf['thal']=minmax_scaling(mf['thal'])
mf['target']=minmax_scaling(mf['target'])
print('Table6 describing mean, standard deviation, min, max and so on in heart disease data base after minmax normalization.')
mf.describe()
```

Based on table 6, the standard deviation decreased significantly. It means that the distribution of data base is between zero and one and we have suppressed big numbers in ths database. it can be considered as a noise suppression method. It is clear that minimum is fixed to zero, and maximum is fixed to 1. 

#### Zscore

First, we need to describe zscore function. It can be described as follow:

```python
def zscore(a):
    xx=a
    a_std = (xx- np.mean(xx)) / (np.std(xx))
    return a_std
```

```python
zf_iris=df_iris_org.copy()
zf_iris['sepal_length']=zscore(zf_iris['sepal_length'])
zf_iris['sepal_width']=zscore(zf_iris['sepal_width'])
zf_iris['petal_length']=zscore(zf_iris['petal_length'])
zf_iris['petal_width']=zscore(zf_iris['petal_width'])
print('Table7 describing mean, standard deviation, min, max and so on in Iris data base after zscore normalization.')
zf_iris.describe()
```

Based on table 7, the mean of databased is almost zero and standard deviation is 1. It is a kind of noise suppersion.
